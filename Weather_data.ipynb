#BeautifulSoup


This project includes the weather data scraping. I have used python to do this project and the library I have use is the `BeautifulSoup`. 

```python
import pandas as pa
from bs4 import BeautifulSoup
import requests as re

url = "https://www.wrh.noaa.gov/zoa/getobext.php?sid=KOKC"
request = re.get(url)

soup = BeautifulSoup(request.text, "lxml")
table = soup.find_all("div", class_="boxit")
```
